```python
import random
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import math
import cPickle
import re
import itertools
%run 'bible_data.ipynb'
%run 'learn_tools.ipynb'
```

```python
unpicklefrom  = "bible.learned2.pickle"
words, _, _ = bible_as_words_inc_punc()
# load the data:
with open(unpicklefrom) as fp:
    saved = cPickle.load(fp)
    
ebed_arr, vdict, rvdict = saved['embeddings'] , saved['vdict'], saved['rvdict']
vocab_size = len(vdict)
```

```python
class Batcher(object):
    def __init__(self, bsize, tsteps, words):
        self._bsize = bsize
        self._tsteps = tsteps
        self._words = words
        self._cursour = [i * (len(words) // bsize) for i in range(bsize)]
    
    def batch(self):
        batches = []
        for t in range(self._tsteps):
            batches.append([])
            for b in range(self._bsize):
                c = self._cursour[b]
                batches[t].append(rvdict[words[c]])
                self._cursour[b] = (c + 1) % len(self._words)
        return batches


def to_words(arr):
    return ' '.join(vdict[i] for i in arr)
    
# batcher = Batcher(5, 3, words)
# print ([to_words(line) for line in zip(*batcher.batch())])   
# print ([to_words(line) for line in zip(*batcher.batch())])    
# print batcher.batch()
```

```python
rnn_graph = tf.Graph()
cell_size = 32
num_steps = 32
batch_size = 64
hidden_dim = 3


with rnn_graph.as_default():
    embeddings = tf.Variable(ebed_arr, trainable=False, name="embeddings")
    
    train_ids = tf.placeholder(tf.int32, [batch_size, num_steps], name="train_ids")
    train_labels_ids = tf.placeholder(tf.int32, [batch_size, num_steps], name="train_labels_ids")
    inputs_train = tf.nn.embedding_lookup(embeddings, train_ids, name="inputs_train")
    
    predict_ids = tf.placeholder(tf.int32, [1], name="predict_ids")
    inputs_predict = tf.nn.embedding_lookup(embeddings, predict_ids, name="inputs_predict")
    
    wo = tf.Variable(tf.random_normal([cell_size, vocab_size], 0.0, 0.1), name="weight")
    bo = tf.Variable(tf.zeros([vocab_size]), name="bias") 
    
    def model(inputs, reuse=False, keep_prop=1.0):
        # L1
        batch_size = inputs[0].get_shape().as_list()[0]
        lstm_cell = tf.nn.rnn_cell.BasicLSTMCell(cell_size, forget_bias=0.0, state_is_tuple=True)
        lstm_cell = tf.nn.rnn_cell.DropoutWrapper(lstm_cell, output_keep_prob=keep_prop)
        cell = tf.nn.rnn_cell.MultiRNNCell([lstm_cell] * hidden_dim, state_is_tuple=True)
        state = cell.zero_state(batch_size, tf.float32)
        state = [[tf.Variable(s, trainable=False, name="state") for s in layer ] for layer in state]
        
        with tf.variable_scope("RNN") as scope:
            if reuse:
                scope.reuse_variables()
            outputs, new_state = tf.nn.rnn(cell, inputs, dtype=tf.float32, initial_state=state)
        
        flat_state = list(itertools.chain.from_iterable(state))
        flat_new_state = list(itertools.chain.from_iterable(new_state))
        
        with tf.control_dependencies([old.assign(new) for old, new in zip(flat_state, flat_new_state)]):
            output = tf.reshape(tf.concat(1, outputs), [-1, cell_size])

        zero_it = tf.group(*[s.assign(tf.zeros_like(s)) for s in flat_state])    
            
        logits = tf.matmul(output, wo) + bo
        return logits, zero_it


    inputs = [tf.squeeze(input_step, [1]) 
              for input_step in tf.split(1, num_steps, inputs_train)] 
    
    logits, clear_train_state = model(inputs, keep_prop=0.8)
    learning_rate = tf.Variable(0.03, trainable=False, name="learning_rate" )
    with tf.control_dependencies([clear_train_state]):
        loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits, tf.reshape(train_labels_ids, [-1])))
        optimzer = tf.train.AdamOptimizer(learning_rate).minimize(loss)
    predict_model, predict_state_clear = model([inputs_predict], reuse=True, keep_prop=1.0)
    prediction = tf.nn.softmax(predict_model)
    saver = tf.train.Saver()
print ("built graph")
```

```python
num_train_steps = 10000
summary = 100
save_at = "biblernn.c32.s32.h3.b64.dropout.ckpt"
load_from = "biblernn.c32.s32.h3.b64.ckpt"
lr = 0.03

with tf.Session(graph=rnn_graph) as sess:
    batcher = Batcher(batch_size, num_steps + 1, words)
    tf.initialize_all_variables().run()
    if load_from:
        saver.restore(sess, load_from)
        print("Loaded from: %s" % load_from)
        
    sess.run([learning_rate.assign(lr)])
    
    for step in range(num_train_steps + 1):
        batch = batcher.batch()
        X = np.transpose(np.array(batch[:-1]))
        Y = np.transpose(np.array(batch[1:]))
        
        _, l = sess.run([optimzer, loss], feed_dict={train_ids:X, train_labels_ids:Y})
        
        if step % summary == 0:
            print(" Step %s, loss %s" % (step, l))
            if save_at:
                save_path = saver.save(sess, save_at)
                print("Model saved in file: %s" % save_path)

            predict_state_clear.run()
            feed = [rvdict['the']]
            sentance = ""
            for i in range(15):
                p = prediction.eval(feed_dict={predict_ids:feed})[0]
                wid = np.random.choice(range(vocab_size), p=p)
                feed = [wid]
                sentance += vdict[wid] + ' '
            print sentance
            print("")
        

    if save_at:
        save_path = saver.save(sess, save_at)
        print("Model saved in file: %s" % save_path)
    
print "Done!"
```
